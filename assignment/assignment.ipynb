```python
import numpy as np # Import the numpy package into your workspace
import matplotlib.pyplot as plt # Import matplotlib into your workspace
import pandas as pd  # Import the pandas package into your workspace
```

1. Cleaning numeric data

```python
air_df = pd.read_csv('data/airbnb_hw.csv', low_memory=False)

# removing the apostrophes so that the strings of prices can be coverted to numeric data
air_df["Price"] = air_df["Price"].str.replace(',', '')
air_df.head()
```

```python
air_df["Price"] = pd.to_numeric(air_df["Price"])
print( air_df["Price"].unique() , '\n')

# converting the prices from strings to numeric for use in plots and operations

print("Number missing: ", sum(air_df["Price"].isnull()))
print("Number of negatives: ", sum(air_df["Price"] < 0))
print("Number of 0s: ", sum(air_df["Price"] == 0))
# none of the prices are null, 0, or negative so nothing else to clean
```

2. Cleaning categorical data

```python
sdf = pd.read_csv("data/sharks.csv", low_memory=False);
sdf.head()
```

```python
shark_type = sdf["Type"]
print(shark_type.value_counts())
```

```python
shark_type = shark_type.replace(['Questionable', 'Unconfirmed', 'Unverified', 'Under investigation', 'Invalid'], np.nan)
# since these categories are all inconclusive, replace with nan

sdf["Type"] = shark_type
print(sdf["Type"].value_counts())
```

```python
shark_type = shark_type.replace(['Boat', 'Watercraft', 'Boating', 'Boatomg'], "Boating")
# these all seem related but are treated as unique types, so combine into one category

sdf["Type"] = shark_type
print(sdf["Type"].value_counts())
```
